## Prewitt filters

**<a id="#">Task 4.1.1.</a>** Load the image `data/astronaut.png` and show it:

```python
img = plt.imread('data/astronaut.png')
plt.imshow(img)
```

**<a id="#">Task 4.1.2.</a>** Implement horizontal and vertical 3&times;3 Prewitt filters:

```python
def prewitt_h(imgp):
    result = numpy.zeros(imgp.shape)
    for p in numpy.ndindex(imgp.shape):
        neighborhood_sum = 0
        if p[0]>=1 and p[0]<imgp.shape[0]-1 and p[1]>=1 and p[1]<imgp.shape[1]-1:
            neighborhood_sum -= sum(imgp[p[0]-1:p[0]+2, p[1]-1]*[1,1,1])
            neighborhood_sum += sum(imgp[p[0]-1:p[0]+2, p[1]+1]*[1,1,1])
            result[p]=neighborhood_sum/6
    return result
    
def prewitt_v(imgp):
    result = numpy.zeros(imgp.shape)
    for p in numpy.ndindex(imgp.shape):
        neighborhood_sum = 0
        if p[0]>=1 and p[0]<imgp.shape[0]-1 and p[1]>=1 and p[1]<imgp.shape[1]-1:
            neighborhood_sum -= sum(imgp[p[0]-1, p[1]-1:p[1]+2]*[1,1,1])
            neighborhood_sum += sum(imgp[p[0]+1, p[1]-1:p[1]+2]*[1,1,1])
            result[p]=neighborhood_sum/6
    return result
   
```

**<a id="#">Task 4.1.3.</a>** Test your above solution using the functions `prewitt_h` and `prewitt_v` for the previously loaded image and show the result:

```python
plt.figure()
plt.imshow(prewitt_h(img))
plt.colorbar()

plt.figure()
plt.imshow(prewitt_v(img))
plt.colorbar()
```

**<a id="#">Task 4.1.4.</a>** Compare your results with the correct result image `data/astronaut_prewitt_h.tiff` and `data/astronaut_prewitt_v.tiff`:

```python
import skimage.io

imgh = skimage.io.imread('data/astronaut_prewitt_h.tiff')
assert numpy.allclose(prewitt_h(img), imgh, atol = 1/255)

imgv = skimage.io.imread('data/astronaut_prewitt_v.tiff')
assert numpy.allclose(prewitt_v(img), imgv, atol = 1/255)
```
